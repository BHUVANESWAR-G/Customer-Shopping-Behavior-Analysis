```python
# 1. Import pandas and load the shopping dataset
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# 2. View the first 5 rows of data
df.head()
```

```python
# 3. Check column types and missing value counts
df.info()
```

```python
# 4. Show summary statistics (mean, max, min, etc.)
df.describe(include='all')
```

```python
# 5. Count how many null values are in each column
df.isnull().sum()
```

```python
# 6. Fill missing Review Ratings using the median of that category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# 7. Confirm there are no more null values
df.isnull().sum()
```

```python
# 8. Clean column names: lowercase, replace spaces with underscores
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# 9. List the new cleaned column names
df.columns
```

```python
# 10. Group ages into 4 categories: Young Adult to Senior
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# 11. Verify the new age groups
df[['age','age_group']].head(10)
```

```python
# 12. Convert text frequencies (like 'Weekly') into number of days (7)
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# 13. Verify the day-mapping results
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# 14. Compare discount and promo code columns
df[['discount_applied','promo_code_used']].head(10)
```

```python
# 15. Check if 'discount_applied' and 'promo_code_used' are identical
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# 16. Remove the duplicate 'promo_code_used' column
df = df.drop('promo_code_used', axis=1)
```

```python
# 17. Final check of all columns before database upload
df.columns
```
